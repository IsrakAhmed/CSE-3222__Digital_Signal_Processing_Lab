```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Unit Step 
n = np.arange(-5,6)
u_n = []

for point in n:
    if point<0:
        u_n.append(0)
    else:
        u_n.append(1)

plt.figure(figsize=(10,15))
plt.subplot(3,2,1)
plt.stem(n,u_n)
plt.title("Unit Step Signal")
plt.xlabel("n")
plt.ylabel("u(n)")


# Ramp
u_r_n = []
for point in n:
    if point<0:
        u_r_n.append(0)
    else:
        u_r_n.append(point)

plt.subplot(3,2,2)
plt.stem(n,u_r_n)
plt.title("Ramp Signal")
plt.xlabel("n")
plt.ylabel("u_r(n)")


# Exponential
x_n = np.exp(n)
plt.subplot(3,2,3)
plt.stem(n,x_n)
plt.title("Exponential Signal")
plt.xlabel("n")
plt.ylabel("x(n)")


# Sine 
t = np.linspace(-5,5,500)
x_t = np.sin(t)
plt.subplot(3,2,4)
plt.plot(t,x_t)
plt.title("Sine Signal")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.grid(True, which='both')
plt.axhline(color='r')
plt.axvline(color='r')


# Cose 
t = np.linspace(-5,5,500)
x_t = np.sin(t)
plt.subplot(3,2,5)
plt.plot(t,x_t)
plt.title("Cos Signal")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.grid(True, which='both')
plt.axhline(color='r')
plt.axvline(color='r')
```
